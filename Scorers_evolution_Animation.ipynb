```python
# Carreguem llibreries
import pandas as pd
import matplotlib.pyplot as plt
import bar_chart_race as bcr
import warnings
warnings.filterwarnings('ignore')
```

```python
#Carreguem les dades
data_raw = pd.read_csv('player_stats.csv')
```

```python
data_raw = data_raw.drop('Unnamed: 0', axis=1)
data_raw
```

```python
# eliminem els jugadors que no tenim dades sobre la divisió en la que jugaren
df = data_raw[data_raw['div'].notna()].reset_index(drop=True)
df
```

```python
#subset de les cols que ens interessen
df_goals = df[['idx','temp','alies','nom', 'G']]
df_goals['G'] = df_goals['G'].astype('int')
df_goals
```

```python
# sumem tos els gols per temporada (hi ha jugadors que han marcat amb més d'un equip la mateixa temporada)
def sum_unique(s):
    return s.unique().sum()


df_goals = df_goals.groupby(['temp','nom'], sort=False, as_index=False).agg(
    idx=('idx','first'),
    temp=('temp', 'first'),
    nom=('nom', 'first'),
    G=('G', sum_unique)
)
df_goals
```

```python
# Trasposem el df i replenem amb 0 els valor nan
df_goals_T = df_goals.pivot_table(index='nom', columns='temp', values='G').T.fillna(0).astype(int)
df_goals_T
```

```python
# eliminem els jugadors que no han marcat cap gol
df_goals_clean = df_goals_T.loc[:, (df_goals_T != 0).any(axis=0)]
df_goals_clean
```

```python
# sumem els gols acumulats al llarg de les temporades
prep_df = df_goals_clean.cumsum()
prep_df
```

```python
# funció que expandeix el dataframe per poder fer la visualització animada

def prepare_data(df, steps=5):
    
    df = df.reset_index()
    df.index = df.index * steps
    last_idx = df.index[-1] + 1
    df_expanded = df.reindex(range(last_idx))
    df_expanded['temp'] = df_expanded['temp'].fillna(method='ffill')
    df_expanded = df_expanded.set_index('temp')
    df_expanded = df_expanded.interpolate().astype(int)
    
    return df_expanded
```

```python
df_exp = prepare_data(prep_df)
df_exp.tail(10)
```
